```python
import pandas as pd
import numpy as np
import os

# ==========================================
# 1. CONFIGURATION
# ==========================================
MY_TEAM_NAME = "Budget Ballers"

# The '~' automatically resolves to C:/Users/YourName/ (Windows) or /Users/YourName/ (Mac)
FOLDER = os.path.expanduser('~/Documents/Python Projects/Fantasy_Baseball_Model/')
TOP_N_FA = 200

print(f"--- STARTING REST-OF-SEASON (ROS) ANALYSIS FOR {MY_TEAM_NAME} ---")
```

```python
# ==========================================
# 2. DATA LOADING & CLEANING
# ==========================================
try:
    df_rosters = pd.read_csv(FOLDER + 'espn_current_rosters.csv', encoding='ISO-8859-1')
    df_hitters = pd.read_csv(FOLDER + 'Fangraphs_Hitter_Projections_ROS.csv', encoding='ISO-8859-1')
    df_pitchers = pd.read_csv(FOLDER + 'Fangraphs_Pitcher_Projections_ROS.csv', encoding='ISO-8859-1')
    df_current = pd.read_csv(FOLDER + 'current_team_stats.csv', encoding='ISO-8859-1')
    print(f"â Success: Projections, Rosters, and Current YTD Stats loaded.")
except Exception as e:
    print(f"X Error: {e}")
    raise

# Clean hidden spaces from team names to ensure they match perfectly
df_rosters['Team'] = df_rosters['Team'].astype(str).str.strip()
df_current['Team'] = df_current['Team'].astype(str).str.strip()

def clean_stat(value):
    if isinstance(value, str) and '%' in value:
        return float(value.replace('%', '')) / 100
    try: return float(value)
    except: return 0.0

for df in [df_rosters, df_hitters, df_pitchers]:
    name_col = 'Player' if 'Player' in df.columns else 'Name'
    df['Clean_Name'] = df[name_col].str.replace(' Jr.', '', regex=False).str.replace(' II', '', regex=False)

hit_stats = ['R', 'HR', 'RBI', 'SB', 'OPS', 'PA']
pit_stats = ['IP', 'QS', 'SV', 'ER', 'H', 'BB']

for col in hit_stats: df_hitters[col] = df_hitters[col].apply(clean_stat)
for col in pit_stats: df_pitchers[col] = df_pitchers[col].apply(clean_stat)
```

```python
# ==========================================
# 3. CORE CALCULATION ENGINE (YTD + ROS)
# ==========================================
def get_league_stats(rosters_df):
    # 1. Calculate Projected Stats (Rest of Season)
    merged_h = pd.merge(rosters_df, df_hitters[['Clean_Name'] + hit_stats], on='Clean_Name', how='left').fillna(0)
    merged_p = pd.merge(rosters_df, df_pitchers[['Clean_Name'] + pit_stats], on='Clean_Name', how='left').fillna(0)
    
    t_hit = merged_h.groupby('Team').agg({'R':'sum','HR':'sum','RBI':'sum','SB':'sum','PA':'sum'}).reset_index()
    merged_h['OPS_num'] = merged_h['OPS'] * merged_h['PA']
    t_hit = pd.merge(t_hit, merged_h.groupby('Team')['OPS_num'].sum().reset_index(), on='Team')
    
    t_pit = merged_p.groupby('Team').agg({'IP':'sum','QS':'sum','SV':'sum','ER':'sum','H':'sum','BB':'sum'}).reset_index()
    
    proj_stats = pd.merge(t_hit, t_pit, on='Team')
    
    # 2. Merge Projections with Current YTD Stats
    stats = pd.merge(df_current, proj_stats, on='Team', suffixes=('_curr', '_proj'))
    
    # 3. Calculate End of Season (EOS) Totals
    # Counting Stats (Simple Addition)
    stats['R'] = stats['R_curr'] + stats['R']
    stats['HR'] = stats['HR_curr'] + stats['HR']
    stats['RBI'] = stats['RBI_curr'] + stats['RBI']
    stats['SB'] = stats['SB_curr'] + stats['SB']
    stats['QS'] = stats['QS_curr'] + stats['QS']
    stats['SV'] = stats['SV_curr'] + stats['SV']
    
    # Ratio Stats (Weighted Math)
    stats['PA'] = stats['PA_curr'] + stats['PA']
    stats['OPS'] = ((stats['OPS_curr'] * stats['PA_curr']) + stats['OPS_num']) / stats['PA'].replace(0, 1)
    
    stats['IP'] = stats['IP_curr'] + stats['IP']
    stats['ER'] = stats['ER_curr'] + stats['ER']
    stats['H'] = stats['H_curr'] + stats['H']
    stats['BB'] = stats['BB_curr'] + stats['BB']
    
    stats['ERA'] = (stats['ER'] / stats['IP'].replace(0, 1)) * 9
    stats['WHIP'] = (stats['H'] + stats['BB'] ) / stats['IP'].replace(0, 1)
    
    # 4. Rank Categories based on true EOS Totals
    for s in ['R', 'HR', 'RBI', 'SB', 'OPS', 'IP', 'QS', 'SV']:
        stats[f'{s}_Pts'] = stats[s].rank(method='min')
    stats['ERA_Pts'] = stats['ERA'].rank(method='min', ascending=False)
    stats['WHIP_Pts'] = stats['WHIP'].rank(method='min', ascending=False)
    
    stats['Total_Points'] = stats[[c for c in stats.columns if '_Pts' in c]].sum(axis=1)
    return stats
```

```python
# ==========================================
# 4. IL-FILTERING LOGIC
# ==========================================
il_slots = ['IL', 'IR', 'Injured', 'Injured Reserve']
active_roster_full = df_rosters[(df_rosters['Status'] == 'Rostered') & 
                                (~df_rosters['Lineup_Slot'].str.contains('|'.join(il_slots), na=False))].copy()

removed_count = len(df_rosters[df_rosters['Status'] == 'Rostered']) - len(active_roster_full)
print(f"â¹ Removed {removed_count} IL players from active calculation.")
```

```python
# ==========================================
# 5. PHASE 1: END OF SEASON STANDINGS
# ==========================================
baseline_stats = get_league_stats(active_roster_full)
display_cols = ['Team', 'Total_Points', 'R', 'HR', 'RBI', 'SB', 'OPS', 'IP', 'QS', 'SV', 'ERA', 'WHIP']
standings = baseline_stats[display_cols].sort_values(by='Total_Points', ascending=False).reset_index(drop=True)
standings.index += 1

int_cols = ['R', 'HR', 'RBI', 'SB', 'IP', 'QS', 'SV']
standings_display = standings.copy()
standings_display[int_cols] = standings_display[int_cols].round(0).astype(int)

print("\n--- 1. PROJECTED END OF SEASON STANDINGS (YTD + ROS) ---")
display(standings_display.style.format({'Total_Points': '{:.1f}', 'OPS': '{:.3f}', 'ERA': '{:.2f}', 'WHIP': '{:.2f}'}))
```

```python
# ==========================================
# 6. PHASE 2: OPTIMIZED SWAPS
# ==========================================
base_team = baseline_stats.set_index('Team').loc[MY_TEAM_NAME]
my_players = active_roster_full[active_roster_full['Team'] == MY_TEAM_NAME].copy()
fa_list = df_rosters[df_rosters['Status'] == 'Available']['Clean_Name'].tolist()

top_fa_h = df_hitters[df_hitters['Clean_Name'].isin(fa_list)].sort_values(by='PA', ascending=False).head(TOP_N_FA)
top_fa_p = df_pitchers[df_pitchers['Clean_Name'].isin(fa_list)].sort_values(by='IP', ascending=False).head(TOP_N_FA)

pt_cols = [c for c in baseline_stats.columns if '_Pts' in c]

def get_impact_string(old_row, new_row):
    impacts = []
    for col in pt_cols:
        diff = new_row[col] - old_row[col]
        if diff != 0:
            label = col.replace('_Pts', '')
            impacts.append(f"{'+' if diff > 0 else ''}{int(diff)} {label}")
    return ", ".join(impacts) if impacts else "Lateral move"

h_results, p_results = [], []

print("\nSimulating best possible swaps against EOS Standings...")

for _, drop in my_players[my_players['Clean_Name'].isin(df_hitters['Clean_Name'])].iterrows():
    for _, add in top_fa_h.iterrows():
        df_temp = active_roster_full.copy()
        df_temp = df_temp[df_temp['Clean_Name'] != drop['Clean_Name']]
        df_temp = pd.concat([df_temp, pd.DataFrame([{'Team': MY_TEAM_NAME, 'Clean_Name': add['Clean_Name'], 'Status': 'Rostered', 'Lineup_Slot': 'BE'}])], ignore_index=True)
        new_stats = get_league_stats(df_temp).set_index('Team').loc[MY_TEAM_NAME]
        gain = new_stats['Total_Points'] - base_team['Total_Points']
        if gain > 0:
            h_results.append({'Add': add['Clean_Name'], 'Drop': drop['Clean_Name'], 'Net Gain': gain, 'Details': get_impact_string(base_team, new_stats)})

for _, drop in my_players[my_players['Clean_Name'].isin(df_pitchers['Clean_Name'])].iterrows():
    is_rp_only = 'RP' in str(drop['Positions']) and 'SP' not in str(drop['Positions'])
    for _, add in top_fa_p.iterrows():
        fa_pos = df_rosters[df_rosters['Clean_Name'] == add['Clean_Name']]['Positions'].values
        fa_pos_str = str(fa_pos[0]) if len(fa_pos) > 0 else ""
        if is_rp_only and 'RP' not in fa_pos_str: continue 
        
        df_temp = active_roster_full.copy()
        df_temp = df_temp[df_temp['Clean_Name'] != drop['Clean_Name']]
        df_temp = pd.concat([df_temp, pd.DataFrame([{'Team': MY_TEAM_NAME, 'Clean_Name': add['Clean_Name'], 'Status': 'Rostered', 'Lineup_Slot': 'BE'}])], ignore_index=True)
        new_stats = get_league_stats(df_temp).set_index('Team').loc[MY_TEAM_NAME]
        gain = new_stats['Total_Points'] - base_team['Total_Points']
        if gain > 0:
            p_results.append({'Add': add['Clean_Name'], 'Drop': drop['Clean_Name'], 'Net Gain': gain, 'Details': get_impact_string(base_team, new_stats)})

print("\n--- 2. TOP 10 HITTER SWAPS ---")
if h_results:
    display(pd.DataFrame(h_results).sort_values(by='Net Gain', ascending=False).drop_duplicates(subset=['Add']).head(10))
else: print("No hitting swaps found.")

print("\n--- 3. TOP 10 PITCHER SWAPS ---")
if p_results:
    display(pd.DataFrame(p_results).sort_values(by='Net Gain', ascending=False).drop_duplicates(subset=['Add']).head(10))
else: print("No pitching swaps found.")
```

```python
# ==========================================
# 7. PHASE 3: MANUAL TRADE & SWAP EVALUATOR
# ==========================================

# --- 1. ENTER YOUR PROPOSED MOVES HERE ---
# You can list multiple players in these brackets! e.g. ["Max Muncy", "Josh Lowe"]
PLAYERS_TO_ACQUIRE = ["Brendan Donovan"] 
PLAYERS_TO_DROP_OR_TRADE = ["Jorge Polanco"]

# If trading with another team, enter their exact team name here (e.g., "Corbin Barrels"). 
# If just picking up a Free Agent, leave this as None.
TRADE_PARTNER = None 

# -----------------------------------------

print(f"\n--- EVALUATING CUSTOM SWAP ---")
print(f"Acquiring: {', '.join(PLAYERS_TO_ACQUIRE)}")
print(f"Shipping Out: {', '.join(PLAYERS_TO_DROP_OR_TRADE)}")
if TRADE_PARTNER: print(f"Trade Partner: {TRADE_PARTNER}")

# Create a fresh copy of the active roster for the simulation
df_sim = active_roster_full.copy()

# 2. Process the players we are giving up
for player in PLAYERS_TO_DROP_OR_TRADE:
    if player in df_sim['Clean_Name'].values:
        if TRADE_PARTNER:
            # Move them to the trade partner's roster
            df_sim.loc[df_sim['Clean_Name'] == player, 'Team'] = TRADE_PARTNER
        else:
            # Remove them entirely (Drop to Free Agency)
            df_sim = df_sim[df_sim['Clean_Name'] != player]
    else:
        print(f"â ï¸ Warning: Could not find '{player}' on your active roster.")

# 3. Process the players we are acquiring
for player in PLAYERS_TO_ACQUIRE:
    # If they are already on a roster (like in a trade), reassign them to your team
    if player in df_sim['Clean_Name'].values:
        df_sim.loc[df_sim['Clean_Name'] == player, 'Team'] = MY_TEAM_NAME
    else:
        # If they are a Free Agent not currently in the active dataframe, append them
        new_player_row = pd.DataFrame([{'Team': MY_TEAM_NAME, 'Clean_Name': player, 'Status': 'Rostered', 'Lineup_Slot': 'BE'}])
        df_sim = pd.concat([df_sim, new_player_row], ignore_index=True)

# 4. Run the simulation engine
sim_stats = get_league_stats(df_sim)

# 5. Extract Before & After for My Team
base_my_team = baseline_stats.set_index('Team').loc[MY_TEAM_NAME]
sim_my_team = sim_stats.set_index('Team').loc[MY_TEAM_NAME]

gain = sim_my_team['Total_Points'] - base_my_team['Total_Points']
impact_str = get_impact_string(base_my_team, sim_my_team)

# 6. Display Results
print(f"\n[ RESULTS FOR {MY_TEAM_NAME} ]")
print(f"Net Point Change: {'+' if gain > 0 else ''}{gain:.1f} Points")
print(f"Category Shifts:  {impact_str}")

# Show a quick preview of the new Top 5 to see if standings shifted
print("\n--- NEW STANDINGS PREVIEW (TOP 5) ---")
display_cols = ['Team', 'Total_Points', 'R', 'HR', 'RBI', 'SB', 'OPS', 'IP', 'QS', 'SV', 'ERA', 'WHIP']
sim_standings = sim_stats[display_cols].sort_values(by='Total_Points', ascending=False).reset_index(drop=True)
sim_standings.index += 1

int_cols = ['R', 'HR', 'RBI', 'SB', 'IP', 'QS', 'SV']
sim_standings_display = sim_standings.copy()
sim_standings_display[int_cols] = sim_standings_display[int_cols].round(0).astype(int)

display(sim_standings_display.head(5).style.format({'Total_Points': '{:.1f}', 'OPS': '{:.3f}', 'ERA': '{:.2f}', 'WHIP': '{:.2f}'}))
```

```python
# ==========================================
# 8. PHASE 4: EMPTY SLOT FILLER (BEST FA BY POSITION)
# ==========================================
TARGET_POSITION = "RP"  # Change this to whatever position you need (e.g., 'OF', 'SP', '1B')
MAX_CANDIDATES_TO_TEST = 50  # Limits how many FAs we simulate to save compute time
TOP_N_RESULTS = 10

print(f"\n--- FINDING BEST AVAILABLE '{TARGET_POSITION}' FOR {MY_TEAM_NAME} ---")

# 1. Identify if we are looking for a Hitter or Pitcher
pitching_positions = ['SP', 'RP', 'P']
is_pitcher = TARGET_POSITION in pitching_positions

# 2. Filter Free Agents by Position
available_fas = df_rosters[df_rosters['Status'] == 'Available'].copy()

# Use regex word boundaries (\b) to ensure we exactly match 'RP' and not something else
available_fas = available_fas[available_fas['Positions'].astype(str).str.contains(rf'\b{TARGET_POSITION}\b', regex=True, na=False)]
fa_names = available_fas['Clean_Name'].tolist()

# 3. Sort by baseline projections so we don't waste time simulating 1,000 terrible players
if is_pitcher:
    candidates = df_pitchers[df_pitchers['Clean_Name'].isin(fa_names)].sort_values(by='IP', ascending=False).head(MAX_CANDIDATES_TO_TEST)
else:
    candidates = df_hitters[df_hitters['Clean_Name'].isin(fa_names)].sort_values(by='PA', ascending=False).head(MAX_CANDIDATES_TO_TEST)

# 4. Run the simulations
fill_results = []
base_my_team = baseline_stats.set_index('Team').loc[MY_TEAM_NAME]

print(f"Simulating adding the top {len(candidates)} projected {TARGET_POSITION}s to your roster...")

for _, add_player in candidates.iterrows():
    df_temp = active_roster_full.copy()
    
    # Add the Free Agent to your team (No drop required)
    new_row = pd.DataFrame([{'Team': MY_TEAM_NAME, 'Clean_Name': add_player['Clean_Name'], 'Status': 'Rostered', 'Lineup_Slot': TARGET_POSITION}])
    df_temp = pd.concat([df_temp, new_row], ignore_index=True)
    
    # Recalculate standings
    new_stats = get_league_stats(df_temp).set_index('Team').loc[MY_TEAM_NAME]
    gain = new_stats['Total_Points'] - base_my_team['Total_Points']
    
    fill_results.append({
        'Add': add_player['Clean_Name'], 
        'Net Gain': gain, 
        'Details': get_impact_string(base_my_team, new_stats)
    })

# 5. Display Results
if fill_results:
    df_results = pd.DataFrame(fill_results).sort_values(by='Net Gain', ascending=False).head(TOP_N_RESULTS)
    display(df_results)
else:
    print(f"â ï¸ No available {TARGET_POSITION}s found in the free agent pool with projections.")
```
